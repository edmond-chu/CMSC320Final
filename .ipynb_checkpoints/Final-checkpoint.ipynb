### What are classified as Twitter bot accounts?
Twitter bots are automated accounts that are created and controlled automatically by bot software. They could have a range of tasks to perform depending on the functionality in mind by the developing party. These bots are often used to work in tandem, and they often mirror the actions of everyday users - whether it be liking tweets, retweeting, or tweeting on their own.

### Why is it important to classify these Twitter bots?
These bots can be manipulated and used on large scales to manipulate complete platforms and push certain agendas. The function and capability of these bots are extremely dangerous to use especially in the political field, where malicious automation can be used to sway voters and promote certain ideas. 
Twitter has written multiple blogs about bots, one of which can be found here - <a href="https://blog.twitter.com/en_us/topics/company/2020/bot-or-not.html" target="_blank">Twitter Blog "Bot or not?"</a>.
There are definitely helpful functionalities of automated accounts, as Twitter says, but they are focusing on the prohibition of bots that focus on "behavior, not content, and are written in a way that targets the spammy tactics of different people or groups could use to try to manipulate conversations of Twitter"

### The application of Data Science 
Data science has many different applications to this area, one of which is using Machine Learning algorithms in order to classify accounts as bots or not, based off of their top level information that we could gather. With labeled training data, we will be able to use that data to train models based off of the attributes we find to be important and use the following model to classify for these bot accounts in the future.

# Exploratory Analysis and Visualization
Now that we have our data let's make it easier to understand by analyzing and then visualizing the imported data. The first things that we find are the mean and median of bots per country to see if there is any correlation between language and number of bots. Median is necessary because the more spoken languages consequently had more bots and we needed to get rid of outliers. We then plot the countries that are in the top 25% of countries to see if there was any large difference. 

```python
num_bots = 0
lang_df = df[df['lang'] != np.nan]
for index, row in df.iterrows():
    if (row['account_type'] == 'bot'):
        num_bots = num_bots + 1

lang_df = lang_df.groupby('lang')
lang_mean = num_bots/(len(lang_df))
print("Average number of bots per language:", lang_mean)

sums = []
for name, group in lang_df:
    sum = 0
    for index, row in group.iterrows():
        if (row['account_type'] == 'bot'):
            sum = sum + 1
    if (sum != 0 ):
        sums.append(sum)

lang_median = np.median(sums)
print("Median number of bots per country:", lang_median)

names = []
bots_per_language = []
for name, group in lang_df:
    sum = 0
    for index, row in group.iterrows():
        if (row['account_type'] == 'bot'):
            sum = sum + 1
    if (sum > lang_median + (lang_median * 0.75) and name != 'en'):
        names.append(name)
        bots_per_language.append(sum)
        
plt.bar(names,bots_per_language)
plt.title('Number of Bots Per Language')
plt.xlabel('Languages')
plt.ylabel('Number of Bots')
plt.show()
```

### Favorites Count
The next parameter that we analyzed was the number of favorites of the twitter users. It is likely that bots will favorite considerably less pictures because they aim to post advertisement and are not concerned with interacting with other users post. Here we create 2 separate dataframes that contain the favorite numbers for bots and humans respectively. The result is as expected and is visualized in the bar graph. This parameter would prove to be useful in our machine learning training.

```python
fav_mean_bot = pd.DataFrame()
fav_mean_bot['favourites'] = df['favourites_count'][df['account_type'] == 'bot']
bot_fav_sum = 0
for index, rows in fav_mean_bot.iterrows():
    bot_fav_sum = bot_fav_sum + int(rows['favourites'])
bot_fav_mean = bot_fav_sum/len(fav_mean_bot)

fav_mean_human = pd.DataFrame()
fav_mean_human['favourites'] = df['favourites_count'][df['account_type'] == 'human']
human_fav_sum = 0
for index, rows in fav_mean_human.iterrows():
    human_fav_sum = human_fav_sum + int(rows['favourites'])
human_fav_mean = human_fav_sum/len(fav_mean_human)

print("Average number of favorites per bot:", bot_fav_mean,"Average number of favorites per human", human_fav_mean)
x_names = ['Bots', 'Humans']
y_values = []
y_values.append(bot_fav_mean)
y_values.append(human_fav_mean)
plt.bar(x_names,y_values)
plt.title('Bot Favorites vs Human Favorites')
plt.xlabel('Account Type')
plt.ylabel('Mean Favorites')
plt.show()
```

### Tweets Per Day
We chose to analyze this data because we reached that there could be two possible conclusions. The first conclusion, would result in bots having more tweets per day because bots have the ability to spam with the same advertisement, while disregarding their content grade. But, users could also have more tweets per day through the use of bots who are used purely to boost certain accounts followers. The results verified the latter conclusion. Humans showed to have almost 2 more tweets per day then bots on average. We use our previous method of splitting into 2 dataframes and then finding the mean of the tweets per day column.

```python
bots_tweets_per_day = pd.DataFrame()
bots_tweets_per_day['tweets_per_day'] = df['average_tweets_per_day'][df['account_type'] == 'bot']
bots_tweets_sum = 0
for index, rows in bots_tweets_per_day.iterrows():
    bots_tweets_sum = bots_tweets_sum + int(rows['tweets_per_day'])
bots_tweets_per_day_mean = bots_tweets_sum/len(bots_tweets_per_day)

human_tweets_per_day = pd.DataFrame()
human_tweets_per_day['tweets_per_day'] = df['average_tweets_per_day'][df['account_type'] == 'human']
human_tweets_sum = 0
for index, rows in human_tweets_per_day.iterrows():
    human_tweets_sum = human_tweets_sum + int(rows['tweets_per_day'])
human_tweets_per_day_mean = human_tweets_sum/len(human_tweets_per_day)

print('Bots mean tweets per day vs Human mean tweets per day:',bots_tweets_per_day_mean, human_tweets_per_day_mean)

fig = plt.figure(figsize =(10, 7))  
ax = fig.add_axes([0, 0, 1, 1]) 
bp = ax.boxplot([bots_tweets_per_day['tweets_per_day'],human_tweets_per_day['tweets_per_day']])
plt.show()
    

print('Bots mean tweets per day vs Human mean tweets per day:',bots_tweets_per_day_mean, human_tweets_per_day_mean)

```

The box and whisker plots show average tweets per day for humans versus bots, as well as visualizes the standard deviations and the outliers.

## Machine Learning Training, Classification, and Analysis.

### Models used
Our end goal of this project is to determine, given the attributes of our dataset, whether or not the account being examined is a bot or a human. With this type of classification, we want classification functions that returns a bit value to correspond whether or not the account is a bot or a human, so we use the SKLearn's Logistic Regression, Support Vector Machine, K Nearest Neighbor, and Random Forest for binary classification.

Below are links to the documentation for each classifier.

## Training and Implementation

To begin, we will use SKLearn's train_test_split function, to split the data into training and testing portions. In this case, we will use 80% of our data for training, 20% for testing.


```python
training_data, testing_data = sklearn.model_selection.train_test_split(param, test_size = .2)
training_y = param['account_type']
training_x = param.drop('account_type',axis = 1)
test_y= param['account_type']
test_x = param.drop('account_type', axis = 1)

```

## Logistic Regression
The below code performs logistic regression training by creating weight vectors for each attribute, and then plotting the observations based off of the given attribute scores for the current account and plots it on a sigmoid function between 0 and 1. Based on the sigmoid function and where the point lies it will have a corresponding probability that tells us whether the account being examined is a bot or not.

Below is an example of what a basic graph of the sigmoid function looks like, as well as the sigmoid function for logistic regression defined as sig(t).

![Graph](https://miro.medium.com/max/2400/1*RqXFpiNGwdiKBWyLJc_E7g.png)

```python
regression = LogisticRegression()
regression.fit(training_x, training_y)
print("Logistic Regression Score = ", regression.score(test_x, test_y))
```

Based off of our output, our accuracy is around 67% in predicing whether the user is a bot or not using Logistic Regression.





## K Nearest Neighbors
K-Nearest Neighbors is another type of classifier that plots points on a plane, and using a distance function(you can use different distance functions, in SKLearn's dependencies they use the euclidean distance), and then find the K nearest neighbors based on the euclidean distance of each point. We will run KNN for different values of K to find the best value of K for classification.

Example of KNN in action is in the graph below, where it shows a new point being classified and calculates the distance from the training instance and compares it to the other nearest neighbors and what they were classified as. If the nearest neighbors are bots, then the new point is likely a bot, and vice versa.
![Graph](https://miro.medium.com/max/650/1*OyYyr9qY-w8RkaRh2TKo0w.png)

```python
neighbors = 1
knn = KNeighborsClassifier(n_neighbors=neighbors)
knn.fit(training_x, training_y)
print("Accuracy of KNN, number of nearest neighbors is =", neighbors, ", ", knn.score(test_x, test_y))
      
neighbors = 2
knn = KNeighborsClassifier(n_neighbors=neighbors)
knn.fit(training_x, training_y)
print("Accuracy of KNN where number of nearest neighbors is =", neighbors, ", ", knn.score(test_x, test_y))

neighbors = 3
knn = KNeighborsClassifier(n_neighbors=neighbors)
knn.fit(training_x, training_y)
print("Accuracy of KNN, number of nearest neighbors is =", neighbors, ", ", knn.score(test_x, test_y))

neighbors = 4
knn = KNeighborsClassifier(n_neighbors=neighbors)
knn.fit(training_x, training_y)
print("Accuracy of KNN, number of nearest neighbors is =", neighbors, ", ", knn.score(test_x, test_y))
      
neighbors = 5
knn = KNeighborsClassifier(n_neighbors=neighbors)
knn.fit(training_x, training_y)
print("K Nearest Neighbors Accuracy where number of nearest neighbors is = ", neighbors, ": ", knn.score(test_x, test_y))
```

So, as shown above, we ran our KNN algorithm where the nearest neighbor ranged from 1-5, with using K = 1 as the best performing classifier at 95% accuracy, and using K = 3 did the second best at 87% accuracy. This is interesting to see K=1 do the best, where it only used one nearest neighbor to classify. This shows there isn't much anomalous data in our data set, since outliers in data could cause problems of only using K=1.



## Support Vector Machines
The next classifier we use is pretty straight forward in its implmentation. How SVM's work is that is a linear classifier using an n-dimensional space that maximizes a plane that defines the distance between our labeled data, bots and humans. When classifying new data, it uses this plot to classify based off its position in the plane.
As shown below, new points that lie on either side of the maximized function are classified as such.
![Graph](https://monkeylearn.com/static/57fd2448dfb67cfff990f32191463e80/6d7b8/plot_hyperplanes_2.png)

```python
svm = SVC(gamma='auto')
svm.fit(training_x, training_y)
print("SVM Accuracy: ", svm.score(test_x, test_y))
```

It looks like our SVM accuracy was 98%, which is our best accuracy so far!


## Random Forests
Random forests is an ensemble learning method for classification, using a multitude of decision trees when training and retuning the result that is the most common among all the decision trees that are created. As shown below, this is a simple random forest where multiple decision trees are created, and the most common result is blue so the classifier returns blue. In our case, we only have to worry about two different outputs, but multiple attributes will contribute to the decisions! To create these decision trees, gini values are calculated for an attributes impurity to create the best possible decision tree. 

![Graph](https://victorzhou.com/media/random-forest-post/random-forest.png)

```python

forest_sqrt = RandomForestClassifier(n_estimators=100, random_state=0, max_features='sqrt')
forest_sqrt.fit(training_x,training_y)
print(forest_sqrt.score(test_x,test_y))
```

After training on 100 different estimators, it looks like random forest performs the best at predicting at a 98.89% accuracy!


## Conclusion 
After running classification using Logistic Regression, K-Nearest Neighbors, Support Vector Machines, and Random Forest, it looks like Random Forest performed the best, classifying at a 99% accuracy.

This result is really good in terms of accuracy, however as shown by running KNN where K=1, it shows that our data doesn't have very anomalous datapoints, so running our model on a larger data set would probably see a decrease in accuracy.
